```python
sns.histplot(df_train.Age, binwidth=5)
```

```python
plt.figure(figsize=(13, 4))

plt.subplot(121)
sns.histplot(df_train[df_train.Exited==0].Age, binwidth=5)

plt.subplot(122)
sns.histplot(df_train[df_train.Exited==1].Age, binwidth=5)
```

```python
plt.figure(figsize=(16, 4))

plt.subplot(131)
sns.histplot(df_train.Age, binwidth=5).set_title('All customers')

plt.subplot(132)
sns.histplot(df_train[df_train.Exited==0].Age, binwidth=5).set_title('Customer does not leave (Exited=0)')

plt.subplot(133)
sns.histplot(df_train[df_train.Exited==1].Age, binwidth=5).set_title('Customer eventually leave (Exited=1)')
```
