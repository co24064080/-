# crawler THSR information
Let's start to crawl the imformation of each station of THSR!
* Target web: [THSR Homepage](https://www.thsrc.com.tw)  
* Choose each station:
    1. Homepage -> Travel Information -> Stations
    2. choose different stations by changing url
* Task:
    1. crawler the name, address, operation hours and ticketing hours of each station
    2. organize into a DataFrame

```python
# setting each station name and their url
stations = ['南港', '台北', '板橋', '桃園', '新竹', '苗栗', '台中', '彰化', '雲林', '嘉義', '台南', '左營']
urls = ['https://www.thsrc.com.tw/ArticleContent/2f940836-cedc-41ef-8e28-c2336ac8fe68',
       'https://www.thsrc.com.tw/ArticleContent/977abb69-413a-4ccf-a109-0272c24fd490',
       'https://www.thsrc.com.tw/ArticleContent/e6e26e66-7dc1-458f-b2f3-71ce65fdc95f',
       'https://www.thsrc.com.tw/ArticleContent/fbd828d8-b1da-4b06-a3bd-680cdca4d2cd',
       'https://www.thsrc.com.tw/ArticleContent/a7a04c89-900b-4798-95a3-c01c455622f4',
       'https://www.thsrc.com.tw/ArticleContent/e8fc2123-2aaf-46ff-ad79-51d4002a1ef3',
       'https://www.thsrc.com.tw/ArticleContent/3301e395-46b8-47aa-aa37-139e15708779',
       'https://www.thsrc.com.tw/ArticleContent/38b8c40b-aef0-4d66-b257-da96ec51620e',
       'https://www.thsrc.com.tw/ArticleContent/5f4c7bb0-c676-4e39-8d3c-f12fc188ee5f',
       'https://www.thsrc.com.tw/ArticleContent/60831846-f0e4-47f6-9b5b-46323ebdcef7',
       'https://www.thsrc.com.tw/ArticleContent/9c5ac6ca-ec89-48f8-aab0-41b738cb1814',
       'https://www.thsrc.com.tw/ArticleContent/f2519629-5973-4d08-913b-479cce78a356']
```

```python
# crawler the address of each station
# crawler the operation hours and ticketing hours of each station

address = []
operation_hours = []
ticketing_hours = []

for i in range(len(stations)):
    response = requests.get(urls[i])
    soup = BeautifulSoup(response.text, 'html.parser')
    
    address.append(soup.find(class_='google-map-link orange').text)
    operation_hours.append(soup.find_all(class_='gray')[0].text[5:])
    ticketing_hours.append(soup.find_all(class_='gray')[1].text[5:])
```

```python
# create a DataFrame
pd.DataFrame({'station':stations, 'address':address, 'operation_hours':operation_hours, 'ticketing_hours':ticketing_hours})
```

# Crawler GDP and CPI
Let's start to crawl GDP and CPI!
* Target web: 
    1. [Wiki GDP](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal))  
    2. [Wiki CPI](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)
* Task:
    1. crawler GDP table (top 50: United States ~ New Zealand)
    2. organize into a DataFrame (columns: Country, Region, IMF_Estimate, IMF_Year, United_Nations_Estimate, United_Nations_Year, World_Bank_Estimate, World_Bank_Year)
    3. crawler CPI table which contains country and 2020 CPI (top 100: Denmark ~ Suriname)
    4. organize into a DataFrame (columns: Country, CPI_2020)
    5. merge GDP(DataFrame) and CPI(DataFrame), based on Country of GDP
    6. plot and text the names of GDP top 10 countrys

```python
# crawler GDP
response = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)')
soup = BeautifulSoup(response.text, 'html.parser')

crawler_data = [[],[],[],[],[],[],[],[]]

for i in range(50):
    A_country = soup.find(class_ = 'wikitable sortable static-row-numbers plainrowheaders srn-white-background').tbody.find_all('tr')[i+2].find_all('td')
    
    for j in range(2):
        crawler_data[j].append(A_country[j].find('a').text)
    for j in range(2,8):
        try:
            A_country[j].find('a').text
            crawler_data[j].append(re.sub('\n', '', re.split(']', A_country[j].text)[1]))
        except:
            crawler_data[j].append(re.sub('\n', '', A_country[j].text))
```

```python
# create GDP DataFrame
GDP = pd.DataFrame({'Country':crawler_data[0], 'Region':crawler_data[1],
              'IMF_Estimate':crawler_data[2], 'IMF_Year':crawler_data[3], 
              'United_Nations_Estimate':crawler_data[4], 'United_Nations_Year':crawler_data[5], 
              'World_Bank_Estimate':crawler_data[6], 'World_Bank_Year':crawler_data[7]})
GDP
```

```python
# crawler CPI
response = requests.get('https://en.wikipedia.org/wiki/Corruption_Perceptions_Index')
soup = BeautifulSoup(response.text, 'html.parser')

crawler_data = [[],[]]
for i in range(2,102):
    A_country = soup.find(class_ = 'wikitable sortable').tbody.find_all('tr')[i].text
    Name_score = re.split('\n', re.split('\xa0', A_country)[-1])
    crawler_data[0].append(Name_score[0])
    crawler_data[1].append(Name_score[1])
```

```python
# CPI DataFrame
CPI = pd.DataFrame({'Country':crawler_data[0], 'CPI_2020':crawler_data[1]})
CPI
```

```python
# merge GDP and CPI DataFrame
GDP_CPI = GDP.merge(CPI, on='Country')
GDP_CPI
```

```python
# plot
Regions = GDP_CPI['Region'].unique()
CPI_2020 = GDP_CPI['CPI_2020'].astype(int)
World_Bank_Estimate = pd.to_numeric(GDP_CPI['World_Bank_Estimate'].str.replace(',', ''))
colors = ['blue', 'orange', 'green', 'red', 'purple']

for i in range(5):
    plt.scatter(CPI_2020[GDP_CPI['Region'] == Regions[i]], World_Bank_Estimate[GDP_CPI['Region'] == Regions[i]], c=colors[i], s=20)
    
plt.xlabel('CPI_2020')
plt.ylabel('World_Bank_Estimate')

for i in range(10):
    plt.text(CPI_2020[i]+0.5, World_Bank_Estimate[i]+1000, GDP_CPI['Country'][i])
    
plt.legend(Regions, loc='center right')
```
